```python
# For demonstration, let's focus on 'total sulfur dioxide' within the 'winequality-red.csv' dataset
df = pd.read_csv('winequality-red.csv')

# --- Added the code here ---
# Step 1: Identify Outliers
Q1 = df['total sulfur dioxide'].quantile(0.25)
Q3 = df['total sulfur dioxide'].quantile(0.75)
# ---------------------------

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Flag rows with 'total sulfur dioxide' outside the interquartile range
outliers_df = df[(df['total sulfur dioxide'] < lower_bound) | (df['total sulfur dioxide'] > upper_bound)]

# Step 2: Investigate Outliers
# Display outliers to understand their characteristics
outliers_df.head()
```

```python
# Compare the statistics of outliers vs. non-outliers
print("Outliers' Statistics:")
outliers_df.describe()
```

```python
print("\nOverall Statistics:")
df.describe()
```

```python
# Step 3: Decide How to Handle Outliers

# Option A: Remove outliers
# --- Added the code here ---
cleaned_df = df[(df['total sulfur dioxide'] >= lower_bound) & (df['total sulfur dioxide'] <= upper_bound)]
# ---------------------------

# Step 4: Re-plot to Verify Changes (for Option A)
plt.figure(figsize=(10, 6))
cleaned_df.boxplot(column='total sulfur dioxide', by='quality')
plt.title('Total Sulfur Dioxide by Wine Quality (Cleaned - Option A)')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Quality')
plt.ylabel('Total Sulfur Dioxide')
plt.show()
```

```python
# Step 3: Decide How to Handle Outliers

# Option B: Cap outliers
# --- Added the code here ---
df['total sulfur dioxide'] = df['total sulfur dioxide'].clip(lower=lower_bound, upper=upper_bound)
# ---------------------------

# Step 4: Re-plot to Verify Changes (for Option B)
plt.figure(figsize=(10, 6))
df.boxplot(column='total sulfur dioxide', by='quality')
plt.title('Total Sulfur Dioxide by Wine Quality (Capped - Option B)')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Quality')
plt.ylabel('Total Sulfur Dioxide')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initial violin plot
fig, ax = plt.subplots()

# --- Added the code here ---
ax.violinplot(dataset=[df_2007[df_2007['continent'] == continent]['life_exp'].values for continent in df_2007['continent'].unique()])
ax.set_xticks(range(1, len(df_2007['continent'].unique()) + 1))
ax.set_xticklabels(df_2007['continent'].unique())
# ---------------------------

ax.set_title('Life Expectancy by Continent in 2007')
ax.set_ylabel('Life Expectancy')

plt.show()
```

```python
# Customizing violin plots
fig, ax = plt.subplots()
continent_names = df_2007['continent'].unique()
dataset = [df_2007[df_2007['continent'] == continent]['life_exp'].values for continent in continent_names]

# Customizing the violin plot parts
# --- Added the code here ---
parts = ax.violinplot(dataset, showmeans=False, showmedians=True, showextrema=True)
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)
# ---------------------------

# Correctly matching quartile markers with continents
for i, continent in enumerate(continent_names, start=1):  # start=1 to match violin plot positions
    quartile1 = df_2007[df_2007['continent'] == continent]['life_exp'].quantile(0.25)
    quartile3 = df_2007[df_2007['continent'] == continent]['life_exp'].quantile(0.75)
    ax.scatter(i, quartile1, marker='_', color='red', s=100, zorder=3)
    ax.scatter(i, quartile3, marker='_', color='blue', s=100, zorder=3)

ax.set_title('Customized Life Expectancy by Continent in 2007')
ax.set_ylabel('Life Expectancy')
ax.set_xticks(range(1, len(continent_names) + 1))
ax.set_xticklabels(continent_names)
plt.xticks(rotation=45)  # Rotate x-tick labels for better readability
plt.show()
```

#### Exercise 13 Integrating Violin Plots with Boxplots

```python
import pandas as pd

df = pd.read_csv('gapminder_full.csv')
df_2007 = df[df['year'] == 2007]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_2007 and life_exp_data are already defined
continents = df_2007['continent'].unique()
life_exp_data = [df_2007[df_2007['continent'] == continent]['life_exp'] for continent in continents]

# Adjust the figsize in the plt.subplots() to increase width
fig, ax = plt.subplots(figsize=(12, 6))  # Increased width from 10 to 12

# --- Added the code here ---
# Create violin plots
parts = ax.violinplot(life_exp_data, positions=np.arange(1, len(continents) * 2, 2), 
                      showmeans=False, showextrema=True, showmedians=True)
# Create boxplots
ax.boxplot(life_exp_data, positions=np.arange(2, len(continents) * 2 + 1, 2), widths=0.5)
# ---------------------------

for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_edgecolor('black')
    pc.set_alpha(0.6)

# Customizing the axes
ax.set_xticks(range(1, len(continents) * 2 + 1, 2))
ax.set_xticklabels(continents)
ax.set_xlim(0, len(continents) * 2 + 1)  # Adjusted the right limit to +1

ax.set_ylabel('Life Expectancy')
ax.set_title('Life Expectancy by Continent in 2007 - Violin and Boxplots')

plt.show()
```

#### Exercise 16 Time Series Analysis with Stack Plots and Trend Line

```python
import pandas as pd

data = {
    'months': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'category_a': [120, 150, 180, 200, 220, 240, 260, 280, 300, 320, 340, 360],
    'category_b': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125, 130, 135],
    'category_c': [40, 42, 44, 46, 48, 50, 52, 54, 56, 58, 60, 62]
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extracting the data
months_idx = range(len(data['months']))  # Numeric x-axis for plotting
categories = ['category_a', 'category_b', 'category_c']
stack_data = [data[cat] for cat in categories]

plt.figure(figsize=(12, 6))

# --- Added the code here ---
# Creating the stack plot
plt.stackplot(months_idx, stack_data, labels=categories, alpha=0.6)
# Calculating total sales for the trend line
total_sales = [sum(x) for x in zip(*stack_data)]
# Fitting a linear trend line to total sales
z = np.polyfit(list(months_idx), total_sales, 1)
p = np.poly1d(z)
# Adding the trend line
plt.plot(list(months_idx), p(list(months_idx)), 'r--', linewidth=2, label='Trend Line')
# ---------------------------

# Customizing the plot
plt.title('Monthly Sales Figures by Product Category with Trend Line')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(ticks=months_idx, labels=data['months'], rotation=45)
plt.legend(loc='upper left')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('cumulative-covid-vaccinations.csv')

df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('cumulative-covid-vaccinations.csv')

# --- Added the code here ---
# Select only World, North America, and Europe
countries_vaccinations = df[df['Entity'].isin(['World', 'North America', 'Europe'])].groupby('Entity')['total_vaccinations'].apply(list)
# ---------------------------

print(countries_vaccinations)
```

```python
cv_dict = countries_vaccinations.to_dict()

#print(cv_dict)
```

```python
max_key, max_value = max(cv_dict.items(), key = lambda x: len(set(x[1])))

cv_dict_full = {}
for k, v in cv_dict.items():
    if len(v) < len(max_value):
        trailing_zeros = [0]*(len(max_value)-len(v))
        cv_dict_full[k] = trailing_zeros + v
    else:
        cv_dict_full[k] = v

#print(cv_dict_full)
```

```python
dates = np.arange(0, len(max_value))

fig, ax = plt.subplots()

# --- Added the code here ---
ax.stackplot(dates, cv_dict_full.values(), labels=cv_dict_full.keys())
# ---------------------------

ax.legend(loc = 'upper left', ncol = 1)
ax.set_title('Cumulative Covid Vaccinations')
ax.set_xlabel('Day')
ax.set_ylabel('Number of People')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('AmesHousing.csv')

# Select only numeric columns for the correlation calculation
numeric_df = df.select_dtypes(include=['number'])

# Calculate the Pearson correlation matrix for numeric columns only
# --- Added the code here ---
pearson_corr = numeric_df.corr()
# ---------------------------

# Create the heatmap
fig, ax = plt.subplots(figsize=(10, 10))
cax = ax.matshow(pearson_corr, cmap='coolwarm')  # Use a diverging color map for clarity

# Add a color bar to provide a reference for the correlation values
fig.colorbar(cax)

# Set the ticks and labels for both axes
plt.xticks(range(len(pearson_corr.columns)), pearson_corr.columns, fontsize=6, rotation=90)
plt.yticks(range(len(pearson_corr.index)), pearson_corr.index, fontsize=6)

# Add text annotations with the correlation values
for i in range(len(pearson_corr.columns)):
    for j in range(len(pearson_corr.index)):
        text = ax.text(j, i, f'{pearson_corr.iloc[i, j]:.1f}',
                       ha="center", va="center", color="w", fontsize=6)

plt.show()
```

#### Exercise 20 Customization of Heatmaps

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('AmesHousing.csv')
```

```python
# Select a subset of variables for simplicity
variables = ['Overall Qual', 'Year Built', 'Total Bsmt SF', 'Gr Liv Area', 'Full Bath', 'Garage Cars', 'SalePrice']
df_subset = df[variables]
df_subset
```

```python
correlation_matrix = df_subset.corr()
correlation_matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(correlation_matrix, cmap='coolwarm')
fig.colorbar(cax)

# Set tick labels
plt.xticks(range(len(variables)), variables, rotation=45, ha='left')
plt.yticks(range(len(variables)), variables)

# Annotate the heatmap with correlation coefficients
# --- Added the code here ---
for i in range(len(variables)):
    for j in range(len(variables)):
        ax.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}',
                ha='center', va='center', color='white', fontsize=10)
# ---------------------------

ax.set_title('Heatmap of Correlation Matrix - Ames Housing Data')
plt.tight_layout()  # Adjust layout to make room for the rotated x-tick labels
plt.show()
```

#### Exercise 21: Time Series Visualization with Heatmaps

```python
import pandas as pd

df = pd.read_csv('AmesHousing.csv')
```

```python
df['Year Sold'] = pd.to_datetime(df['Yr Sold'], format='%Y').dt.year
df['Month Sold'] = df['Mo Sold']
monthly_prices = df.pivot_table(values='SalePrice', index='Year Sold', columns='Month Sold', aggfunc='mean')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(12, 8))

# --- Added the code here ---
cax = ax.matshow(monthly_prices.values, cmap='YlOrRd')
# ---------------------------

fig.colorbar(cax)

# Set x-axis labels to months, y-axis labels to years
ax.set_xticks(np.arange(len(monthly_prices.columns)))
ax.set_yticks(np.arange(len(monthly_prices.index)))
ax.set_xticklabels(monthly_prices.columns)
ax.set_yticklabels(monthly_prices.index)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

ax.set_xlabel('Month')
ax.set_ylabel('Year')
ax.set_title('Average Monthly Sale Price Over Years - Heatmap')

plt.show()
```
