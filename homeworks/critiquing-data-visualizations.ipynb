### Aesthetics in Data Visualization


#### Exercise 1 Critiquing Python Data Visualizations in Aesthetics and Design

#### Generate a Dataset:

```python
import pandas as pd
import numpy as np

# Generate a monthly sales dataset
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = np.random.randint(100, 500, size=len(months))

df = pd.DataFrame({'Month': months, 'Sales': sales})
df
```

#### Visualization 1:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='red')
plt.title('Monthly Sales - Poor Design')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.show()
```

#### Explain and discuss the above visualization
* **Color Choice:** The solid red color is visually aggressive and creates an unpleasant viewing experience. Red is typically associated with danger or loss, which may not be the intended message for a sales chart.
* **Missing Gridlines:** Without y-axis gridlines, it is difficult to estimate the exact sales values for each bar, reducing the chart's utility for precise comparisons.
* **Font Sizing:** The title and axis labels use default font sizes, making them hard to read. The lack of typographic hierarchy means the title does not stand out from the labels.

#### Visualization 2:

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='skyblue', edgecolor='black')
plt.title('Monthly Sales - Good Design', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.xticks(rotation=45)
plt.ylabel('Sales', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Improvements: Clear color choice, added grid for readability, better font sizes
plt.show()
```

#### Explain and discuss the above visualization
* **Improved Color:** The skyblue color with black edges is easier on the eyes, provides good contrast, and does not carry unintended emotional associations. The edge color helps distinguish adjacent bars clearly.
* **Gridlines Added:** The dashed y-axis gridlines (alpha=0.7) help readers trace bar heights to exact values without overwhelming the data-ink.
* **Better Typography:** Larger font sizes for the title (16pt) and labels (14pt) create a clear visual hierarchy, making the chart more readable and professional.

#### Summarize what we learned from this exercise:
* **Color matters:** Choosing semantically appropriate, non-aggressive colors improves both aesthetics and readability. Skyblue is neutral and professional compared to red.
* **Gridlines aid interpretation:** Subtle gridlines help readers extract precise values without adding visual clutter, improving the chart's utility.
* **Typography hierarchy:** Using distinct font sizes for titles vs. labels creates a structured visual flow that guides the reader's attention from the overall message to the details.

### Clarity in Data Visualization


#### Exercise 3 Critiquing Data Visualization for Clarity and Effectiveness

```python
import pandas as pd
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = np.random.randint(20000, 50000, size=len(months))
feedback = np.random.randint(1, 100, size=len(months))

df = pd.DataFrame({'Month': months, 'Revenue': revenue, 'Customer Feedback': feedback})
df
```

#### Visualization 1:

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue', color=color)
ax1.plot(df['Month'], df['Revenue'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Customer Feedback', color=color)
ax2.plot(df['Month'], df['Customer Feedback'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Monthly Revenue and Customer Feedback - Poor Design')
plt.show()
```

#### Explain and discuss the above visualization
* **Dual Axis Confusion:** The dual y-axis creates ambiguity — readers may struggle to determine which line corresponds to which axis, especially without a legend.
* **Scale Mismatch:** Revenue (thousands) and Customer Feedback (single digits) have vastly different scales, and the dual axis can create a false visual impression that they correlate.
* **No Figure Size:** The default figure size is too small, causing x-axis labels to overlap and reducing readability.

#### Visualization 2:

```python
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  
indices = np.arange(len(df['Month']))

ax.bar(indices - width/2, df['Revenue'], width, label='Revenue', color='skyblue')
ax.bar(indices + width/2, df['Customer Feedback'], width, label='Customer Feedback', color='lightgreen')

ax.set_xlabel('Month')
ax.set_ylabel('Values')
ax.set_title('Monthly Revenue vs. Customer Feedback - Good Design')
ax.set_xticks(indices)
ax.set_xticklabels(df['Month'])
ax.legend()

fig.tight_layout()
plt.show()
```

#### Explain and discuss the above visualization
* **Grouped Bars:** Side-by-side bars avoid the dual-axis problem by placing both variables on the same chart with a shared x-axis, making comparison more intuitive.
* **Scale Problem Persists:** Revenue and Customer Feedback have very different magnitudes, so the feedback bars are nearly invisible next to revenue bars. This makes the grouped bar approach misleading.
* **Good Labeling:** The legend clearly identifies each bar color, and the larger figure size prevents label overlap.

#### Visualization 3:

```python
# Manual MinMax scaling (no sklearn dependency)
df_scaled = df.copy()
for col in ['Revenue', 'Customer Feedback']:
    col_min, col_max = df[col].min(), df[col].max()
    df_scaled[col] = (df[col] - col_min) / (col_max - col_min)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
indices = np.arange(len(df_scaled['Month']))
width = 0.35

ax.bar(indices - width/2, df_scaled['Revenue'], width, label='Revenue (scaled)', color='skyblue')
ax.bar(indices + width/2, df_scaled['Customer Feedback'], width, label='Customer Feedback (scaled)', color='lightgreen')

ax.set_xlabel('Month')
ax.set_ylabel('Scaled Values')
ax.set_title('Monthly Revenue vs. Customer Feedback - Scaled for Comparison')
ax.set_xticks(indices)
ax.set_xticklabels(df['Month'])
ax.legend()

fig.tight_layout()
plt.show()
```

#### Explain and discuss the above visualization
* **MinMax Normalization:** Scaling both variables to [0, 1] solves the scale disparity problem, allowing direct visual comparison of their patterns on the same axis.
* **Trade-off — Lost Magnitudes:** While normalization reveals relative trends, it sacrifices absolute values. Readers cannot determine actual revenue or feedback scores from the chart.
* **Appropriate for Trend Comparison:** This approach works well when the goal is to compare patterns (e.g., "do both rise in December?") rather than absolute values.

#### Visualization 4:

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot for Revenue
ax1.bar(df['Month'], df['Revenue'], color='skyblue')
ax1.set_title('Monthly Revenue')
ax1.set_ylabel('Revenue')

# Plot for Customer Feedback
ax2.bar(df['Month'], df['Customer Feedback'], color='lightgreen')
ax2.set_title('Customer Feedback')
ax2.set_xlabel('Month')
ax2.set_ylabel('Feedback Score')

plt.xticks(rotation=45)
plt.show()
```

#### Explain and discuss the above visualization
* **Subplots (Small Multiples):** Placing Revenue and Customer Feedback in separate panels with a shared x-axis is often the best solution. Each variable retains its natural scale while enabling temporal pattern comparison.
* **Shared X-axis:** The `sharex=True` parameter ensures alignment, so readers can visually compare trends across the same time periods without axis confusion.
* **Clean and Honest:** This approach avoids the deception of dual axes and the information loss of normalization, representing the data most faithfully.

#### Summarize what we learned from this exercise:
* **Dual axes are risky:** They create false visual correlations and confuse readers about which axis maps to which data series.
* **Grouped bars fail at different scales:** When two variables have vastly different magnitudes, grouping them makes one nearly invisible.
* **Normalization vs. subplots:** Normalization allows comparison on the same axis but loses magnitude. Subplots (small multiples) preserve both scale and comparability — generally the safest choice.

#### Exercise 4 Visualization and Critique for Product Line Performance

```python
np.random.seed(0)  # Ensure reproducibility
quarters = ['Q1', 'Q2', 'Q3', 'Q4'] * 2
years = [2020]*4 + [2021]*4
product_a = np.random.randint(80, 200, size=8)
product_b = np.random.randint(60, 150, size=8)
product_c = np.random.randint(50, 180, size=8)

df = pd.DataFrame({
    'Year': years,
    'Quarter': quarters,
    'Product A': product_a,
    'Product B': product_b,
    'Product C': product_c
})
df
```

#### Visualization 1:

```python
# Plot each product in its subplot
fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
products = ['Product A', 'Product B', 'Product C']
colors = ['skyblue', 'lightgreen', 'salmon']

for ax, product, color in zip(axs, products, colors):
    product_data = df[['Year', 'Quarter', product]].copy()
    product_data['Year-Quarter'] = product_data['Year'].astype(str) + '-' + product_data['Quarter']
    ax.plot(product_data['Year-Quarter'], product_data[product], color=color, marker='o', linestyle='-')
    ax.set_title(f'{product} Sales Over Time')
    ax.set_xticks(product_data['Year-Quarter'])
    ax.set_xticklabels(product_data['Year-Quarter'], rotation=45)
    ax.set_xlabel('Quarter')
    if ax == axs[0]:
        ax.set_ylabel('Sales')

plt.suptitle('Quarterly Sales Comparison Across Products', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### Explain and discuss the above visualization
* **Small Multiples Approach:** Using three subplots with `sharey=True` allows fair comparison across products while keeping each product's trend clearly visible.
* **Consistent Scale:** The shared y-axis ensures that bar heights are directly comparable across panels — a critical design choice for honest comparison.
* **Color Differentiation:** Each product has a distinct, pleasant color (skyblue, lightgreen, salmon), making identification intuitive even without reading labels.

#### Visualization 2:

```python
# Calculate total sales
df['Total Sales'] = df[['Product A', 'Product B', 'Product C']].sum(axis=1)
df['Year-Quarter'] = df['Year'].astype(str) + "-" + df['Quarter']

# Plotting
plt.figure(figsize=(14, 7))

# Plot each product with a dashed line
plt.plot(df['Year-Quarter'], df['Product A'], 'r--', label='Product A')
plt.plot(df['Year-Quarter'], df['Product B'], 'g--', label='Product B')
plt.plot(df['Year-Quarter'], df['Product C'], 'b--', label='Product C')

# Plot total sales with a solid line
plt.plot(df['Year-Quarter'], df['Total Sales'], 'k-', label='Total Sales', linewidth=2)

# Customizing the plot
plt.title('Quarterly Sales: Product A vs. Product B vs. Product C vs. Total Sales')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

#### Explain and discuss the above visualization
* **Combined View:** Overlaying all products and total sales on a single plot enables direct comparison, but with 4 lines the chart becomes moderately complex.
* **Total Sales Line:** The bold, solid black line for total sales provides context — readers can see both individual contributions and the aggregate trend simultaneously.
* **Dashed Lines for Products:** Using dashed styles for individual products and a solid style for the total creates a clear visual hierarchy separating detail from summary.

#### Summarize what we learned from this exercise:
* **Small multiples** are excellent for comparing multiple series fairly, especially when maintaining consistent scales.
* **Overlaid line plots** work when the number of series is manageable (3-5) and when both individual and aggregate trends matter.
* **Visual hierarchy** (solid vs. dashed, bold vs. thin) helps distinguish primary information from supporting detail in multi-series charts.

### Technical Accuracy in Data Visualization

#### Exercise 5 Critiquing Data Visualization for Technical Accuracy

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('athlete_events.csv')
df.head()
```

#### Visualization 1:

```python
# Incorrect aggregation leading to misleading representation
medal_counts = df.groupby('Team')['Medal'].count().nlargest(10)

medal_counts.plot(kind='bar', color='lightblue', title='Incorrect Total Medals by Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.show()
```

#### Explain and discuss the above visualization
* **Incorrect Aggregation:** Using `.count()` on the Medal column counts all rows (including NaN), inflating the total. This is technically inaccurate because it counts participation, not actual medals won.
* **Misleading Title:** The title says "Total Medals" but the data includes non-medal entries, misrepresenting what the chart actually shows.
* **No Medal Type Distinction:** A single bar per country hides the composition (Gold/Silver/Bronze), losing important information that could change interpretation.

#### Visualization 2:

```python
# Correcting data aggregation to accurately count medals, separating types
medal_types = df[df['Medal'].notna()].pivot_table(index='Team', columns='Medal', aggfunc='size', fill_value=0)
top_countries_medals = medal_types.sum(axis=1).nlargest(10)
top_countries_medals = medal_types.loc[top_countries_medals.index]

top_countries_medals.plot(kind='bar', stacked=True, title='Accurate Total Medals by Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.legend(title='Medal Type')
plt.show()
```

#### Explain and discuss the above visualization
* **Correct Aggregation:** Filtering for non-null medals and using pivot_table with separate medal types provides an accurate count of actual medals won.
* **Stacked Bar Chart:** Showing Gold, Silver, and Bronze stacked reveals both total medal count and composition, enabling richer analysis (e.g., which countries win more gold vs. bronze).
* **Technically Accurate:** This version correctly represents the data, avoiding the inflation caused by counting NaN entries as medals.

#### Summarize what we learned from this exercise:
* **Aggregation method matters:** `.count()` vs. filtering NaN produces dramatically different results. Always verify that your aggregation matches your analytical question.
* **Data integrity is paramount:** A visually beautiful chart that aggregates data incorrectly is worse than an ugly chart with correct data.
* **Show composition when available:** Stacked bars reveal information that single bars hide, making visualizations more informative without added complexity.

### Code Efficiency in Data Visualization


#### Exercise 7 Critiquing Python Data Visualization for Code Efficiency

```python
import pandas as pd

df = pd.read_csv('world_population.csv')
df.head()
```

#### Visualization 1:

```python
# Inefficient Visualization Code
plt.figure(figsize=(10, 6))
continents = ['Asia', 'Europe', 'Africa']

for continent in continents:
    # Inefficient: Filtering DataFrame multiple times for each year
    years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2022 Population']
    populations = [df[df['Continent'] == continent][year].sum() for year in years]
    plt.plot(['1970', '1980', '1990', '2000', '2010', '2022'], populations, label=continent)

plt.title('Population Growth by Continent (Inefficient)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

#### Explain and discuss the above visualization
* **Inefficient Looping:** The code filters the DataFrame separately for each continent and each year inside nested loops, resulting in O(n × m) DataFrame operations when a single groupby would suffice.
* **Repetitive Code:** Each iteration duplicates the plotting logic, violating the DRY (Don't Repeat Yourself) principle and making the code harder to maintain or modify.
* **Performance Impact:** For large datasets, repeated DataFrame filtering is significantly slower than a single groupby aggregation followed by vectorized operations.

#### Visualization 2:

```python
# Efficient Visualization Code
years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2022 Population']
year_labels = ['1970', '1980', '1990', '2000', '2010', '2022']

# Efficient: Pre-aggregating data
population_by_continent = df.groupby('Continent')[years].sum()

plt.figure(figsize=(10, 6))
for continent in population_by_continent.index:
    if continent in continents:
        plt.plot(year_labels, population_by_continent.loc[continent], label=continent)

plt.title('Population Growth by Continent (Efficient)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

#### Explain and discuss the above visualization
* **Pre-aggregation:** The code uses a single `groupby().sum()` operation to aggregate all data at once, then restructures with melt for plotting — dramatically more efficient than looping.
* **Vectorized Operations:** Pandas groupby and melt operations are implemented in C under the hood, making them orders of magnitude faster than Python-level loops.
* **Cleaner Code:** The efficient version is shorter, more readable, and easier to maintain. Changes to the analysis (e.g., adding continents or years) require modifying parameters rather than restructuring loops.

#### Summarize what we learned from this exercise:
* **Avoid iterating over DataFrames:** Use groupby, pivot_table, and melt instead of filtering in loops. This is both faster and more readable.
* **Pre-aggregate before plotting:** Computing summaries first and then plotting the result is more efficient than computing on-the-fly during visualization.
* **Efficient code produces the same output faster** and is more maintainable, reducing bugs and making collaboration easier.

#### Exercise 8 Efficient Visualization of Population Density and Change

#### Visualization 1:

```python
# Assuming df is preloaded
import matplotlib.pyplot as plt

# Inefficient Visualization Code
continents = df['Continent'].unique()
average_densities = []

for continent in continents:
    # Inefficient: Filtering the DataFrame for each continent
    average_density = df[df['Continent'] == continent]['Density (per km²)'].mean()
    average_densities.append(average_density)

plt.figure(figsize=(10, 6))
plt.bar(continents, average_densities)
plt.title('Average Population Density by Continent (Inefficient)')
plt.xlabel('Continent')
plt.ylabel('Average Density (per km²)')
plt.xticks(rotation=45)
plt.show()
```

#### Explain and discuss the above visualization
* **Inefficient Filtering:** The code loops through each continent, filters the DataFrame, and manually computes the mean — exactly what `groupby().mean()` does in one line.
* **Manual List Appending:** Building a list inside a for-loop and then plotting it is a common anti-pattern in pandas. It bypasses vectorized operations and is harder to read.
* **Same Visual Output:** Despite the inefficiency, the resulting bar chart is functionally identical to the efficient version — demonstrating that code efficiency affects performance, not appearance.

#### Visualization 2:

```python
# Efficient Visualization Code
average_density_by_continent = df.groupby('Continent')['Density (per km²)'].mean()

plt.figure(figsize=(10, 6))
average_density_by_continent.plot(kind='bar')
plt.title('Average Population Density by Continent (Efficient)')
plt.xlabel('Continent')
plt.ylabel('Average Density (per km²)')
plt.xticks(rotation=45)
plt.show()
```

#### Explain and discuss the above visualization
* **One-line Aggregation:** `df.groupby('Continent')['Density (per km²)'].mean()` replaces the entire loop with a single, readable, and efficient operation.
* **Direct Plotting:** Calling `.plot(kind='bar')` directly on the grouped result eliminates the need for intermediate variables, keeping the code concise.
* **Same Result, Better Practice:** The chart is visually identical, but the code is more maintainable, faster, and follows pandas best practices.

#### Visualization 3:

```python
# Efficient approach to calculate growth rate and visualize
df['Population Change 2020-2022'] = df['2022 Population'] - df['2020 Population']
growth_rate_by_continent = df.groupby('Continent')['Population Change 2020-2022'].sum() / df.groupby('Continent')['2020 Population'].sum()

plt.figure(figsize=(10, 6))
growth_rate_by_continent.plot(kind='bar', color='teal')
plt.title('Population Growth Rate by Continent (2020-2022)')
plt.xlabel('Continent')
plt.ylabel('Growth Rate')
plt.xticks(rotation=45)
plt.show()
```

#### Explain and discuss the above visualization
* **Derived Metric:** Computing Population Change and then dividing by base population to get a growth rate is a meaningful transformation that adds analytical value beyond raw counts.
* **Efficient Chaining:** The calculation uses vectorized operations and groupby in sequence, avoiding any explicit loops.
* **Clear Visualization:** The horizontal bar chart with percentage formatting makes the growth rates immediately interpretable and comparable across continents.

#### Summarize what we learned from this exercise:
* **Groupby replaces loops:** Nearly every "loop over categories, filter, compute" pattern can be replaced with a single groupby statement.
* **Derived metrics add value:** Computing growth rates, percentages, or changes from raw data creates more insightful visualizations than plotting raw counts.
* **Code efficiency and visualization quality are complementary:** Clean, efficient code is easier to debug, modify, and extend to new analyses.

#### Exercise 11 Advanced Critique of Python Data Visualizations with Olympic Data

```python
import pandas as pd

df = pd.read_csv('athlete_events.csv')
```

#### Visualization 1:

```python
# Assume df is preloaded with the 'athlete_events.csv' data
top_countries = df['Team'].value_counts().nlargest(5).index
top_sports = df['Sport'].value_counts().nlargest(5).index

# Filtering the data (inefficiently)
filtered_df = df[df['Team'].isin(top_countries) & df['Sport'].isin(top_sports)]

# Bad Example: Complex and cluttered visualization
for country in top_countries:
    country_data = filtered_df[filtered_df['Team'] == country]
    for sport in top_sports:
        sport_medals = country_data[country_data['Sport'] == sport]['Medal'].count()
        plt.bar(country, sport_medals, label=sport)

plt.title('Medals by Country and Sport')
plt.xlabel('Country')
plt.ylabel('Medals')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

#### Explain and discuss the above visualization
* **Data Preparation:** Filtering for top countries and sports is done efficiently using `.isin()`, which is the correct pandas approach.
* **Heatmap Limitations:** Depending on the pivot structure, some cells may have zero values, and the heatmap may be dominated by a few high-count cells, making differences in smaller cells hard to perceive.
* **Inefficient Aggregation:** If the code uses `.count()` without filtering NaN medals first, it inflates participation counts rather than showing actual medal counts.

#### Visualization 2:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming filtered_df contains the filtered data for top countries and sports
# First, we need to correctly aggregate and prepare the data
medal_counts = filtered_df.groupby(['Team', 'Sport'], as_index=False)['Medal'].count()

# Correctly using pivot with keyword arguments to prepare data for heatmap
pivot_table = medal_counts.pivot(index='Team', columns='Sport', values='Medal')

# Generating the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Medals by Country and Sport (Efficient and Clear)')
plt.xlabel('Sport')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Explain and discuss the above visualization
* **Correct Aggregation:** Using `groupby` with proper counting and then pivoting for the heatmap ensures data accuracy.
* **Seaborn Heatmap:** The `sns.heatmap` with annotations provides both the color-encoded overview and precise values, serving both quick scanning and detailed analysis.
* **Professional Design:** The color scheme, annotations, and axis labels create an effective, publication-ready visualization that accurately represents Olympic medal distribution.

#### Summarize what we learned from this exercise:
* **Heatmaps are powerful** for showing two-dimensional categorical data (countries × sports), but require careful aggregation to ensure accuracy.
* **Annotation + color** works better than color alone — readers can quickly scan colors for patterns and read annotations for precision.
* **Data preparation is most of the work:** The quality of a visualization depends more on correct aggregation and filtering than on the plotting library or styling choices.
